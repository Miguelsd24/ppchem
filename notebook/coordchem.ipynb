```python
import py3Dmol
import os

# =========================
# LOAD SDF
# =========================
path = os.path.join("../src", "complex.sdf")

with open(path) as f:
    data = f.read()

# =========================
# VIEW
# =========================
view = py3Dmol.view(width=700, height=550)
view.addModel(data, "sdf")

# =========================
# STICKS (BONDS)
# =========================
view.setStyle({}, {
    "stick": {
        "radius": 0.22,
        "color": "gray",
        "opacity": 0.9
    }
})

# =========================
# ATOMS (SPHERES REDUCED SIZE)
# =========================
view.addStyle({"elem": "Fe"}, {
    "sphere": {
        "scale": 0.75,
        "color": "red"
    }
})

view.addStyle({"elem": "N"}, {
    "sphere": {
        "scale": 0.35,
        "color": "blue"
    }
})

view.addStyle({"elem": "H"}, {
    "sphere": {
        "scale": 0.22,
        "color": "white"
    }
})

# =========================
# CAMERA
# =========================
view.zoomTo()
view.show()
```
